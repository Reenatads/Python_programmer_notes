```python
# Criando gráfico de dispersão

plt.scatter(df.age, df.height)

# Editando o gráfico

plt.scatter(df.age, df.height,
color='green',
marker='s')

# Ajustando a transparência

plt.scatter(df.x_data,
df.y_data,
alpha=0.1)
```

```python
# Criando um gráfico de barras

plt.bar(df.precinct,
df.pets_abducted)

plt.ylabel('Pet Abductions')
plt.show()

# Criando gráfico de barras na horizontal

plt.barh(df.precinct,
df.pets_abducted)

plt.ylabel('Pet Abductions')
plt.show()

# Adicionando barra de erros

plt.bar(df.precinct, df.pet_abductions,
yerr=df.error)
plt.ylabel('Pet Abductions')
plt.show()

# Stacked bar chart, gráfico com duam informações em uma barra.

```
